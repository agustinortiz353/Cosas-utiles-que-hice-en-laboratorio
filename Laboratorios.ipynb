```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
def next_even(n):
    """Devuelve el siguiente número par a partir de un número dado."""
    return n + 2 if n % 2 == 0 else n + 1


# Define las coordenadas X e Y para la cuadrícula
x_min, x_max = -10, 10
y_min, y_max = -10, 10
step = 0.1
X, Y = np.meshgrid(np.arange(x_min, x_max, step), np.arange(y_min, y_max, step))

# Puntos pertenecientes a las cinco curvas equipotenciales
points = {
    "V1": [(-5, 0), (-5.1, -1), (-5.4, -2), (-5.9, -3), (-6.7, -4), (-5.1, 1), (-5.2, 2), (-5.7, 3)],
    "V2": [(-3, 0), (-3, -1), (-3.1, -2), (-3.3, -3), (-3.6, -4), (-3.9, -5), (-3, 1), (-3.1, 2), (-3.3, 3), (-3.6, 4)],
    "V3": [(-1, 0), (-1, -1), (-1.1, -2), (-1.1, -3), (-1.2, -4), (-1.3, -5), (-1, 1), (-1, 2), (-1.2, 3), (-1.4, 4), (-1.4, 5), (-1.6, 6)],
    "V4": [(1, 0), (1, -1), (1, -2), (1, -3), (1.1, -4), (1.2, -5), (1.3, -6), (1, 1), (1, 2), (1.1, 3), (1.1, 4), (1.3, 5), (1.5, 6), (1.7, 7)],
    "V5": [(2.5, 0), (2.6, -1), (2.6, -2), (2.7, -3), (2.8, -4), (3.1, -5), (4, -6), (2.7, 1), (2.7, 2), (2.8, 3), (2.9, 4), (3.2, 5), (4, 6)]
}

# Función para calcular el campo eléctrico debido a dos barras cargadas linealmente con carga opuesta
def electric_field_two_lines_opposite(length1, width1, r1, length2, width2, r2, r):
    """Calcula el campo eléctrico debido a dos barras cargadas linealmente con carga opuesta, una con longitud length1, ancho width1
    y posición r1, y la otra con longitud length2, ancho width2 y posición r2, en el punto r."""
    k = 8.9875517923 * 10**9  # Constante de Coulomb en N m^2 / C^2
    delta_r1 = r - r1
    r1_mag = np.linalg.norm(delta_r1)
    delta_r2 = r - r2
    r2_mag = np.linalg.norm(delta_r2)
    E1 = k * width1 * length1 * delta_r1 / r1_mag**3
    E2 = - k * width2 * length2 * delta_r2 / r2_mag**3  # Carga opuesta
    return E1 + E2

# Definir las dimensiones y posiciones de las barras
length1 = 2  # Longitud de la primera barra
width1 = 8   # Ancho de la primera barra
r1 = np.array([-6.8, 0])  # Posición de la primera barra
length2 = 2  # Longitud de la segunda barra
width2 = 12   # Ancho de la segunda barra
r2 = np.array([4, 0])   # Posición de la segunda barra

# Calcular el campo eléctrico en cada punto de la cuadrícula
Ex = np.zeros_like(X)
Ey = np.zeros_like(Y)
for i, row in enumerate(X):
    for j, col in enumerate(row):
        r = np.array([X[i, j], Y[i, j]])
        field = electric_field_two_lines_opposite(length1, width1, r1, length2, width2, r2, r)
        Ex[i, j] = field[0]
        Ey[i, j] = field[1]

# Trazar las líneas de campo eléctrico y las curvas equipotenciales en el mismo gráfico
plt.figure(figsize=(8, 6))

# Trazar las líneas de campo eléctrico
plt.streamplot(X, Y, Ex, Ey, color='blue', density=1)

# Trazar las curvas equipotenciales
for label, pts in points.items():
    x_vals, y_vals = zip(*pts)
    plt.scatter(x_vals, y_vals, label=label)
    degree = next_even(len(pts) // 2)
    poly = np.poly1d(np.polyfit(y_vals, x_vals, degree))
    y_fit = np.linspace(min(y_vals), max(y_vals), 1000)
    x_fit = poly(y_fit)
    plt.plot(x_fit, y_fit, '--', color='gray')

# Trazar las barras
plt.plot([r1[0], r1[0]], [r1[1] - width1 / 2, r1[1] + width1 / 2], color='red', linewidth=2, label='Barra 1')
plt.plot([r2[0], r2[0]], [r2[1] - width2 / 2, r2[1] + width2 / 2], color='green', linewidth=2, label='Barra 2')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Curvas Equipotenciales con Líneas de Campo Eléctrico')
plt.grid(True)
plt.legend()
plt.show()

# Calcular las incertezas de los coeficientes del ajuste polinomial para cada curva equipotencial
for label, pts in points.items():
    print(f"Incertezas del ajuste para la curva equipotencial {label}:")
    x_vals, y_vals = zip(*pts)
    degree = next_even(len(pts) // 2)
    poly_coeffs, poly_cov = np.polyfit(y_vals, x_vals, degree, cov=True)
    poly_fit = np.poly1d(poly_coeffs)
    print(f"Curva Equipotencial {label}:")
    print(poly_fit)
    for i, coeff in enumerate(poly_coeffs):
        uncertainty = np.sqrt(poly_cov[i, i])
        # Truncar la incerteza para que coincida con la cantidad de cifras significativas del coeficiente
        uncertainty_truncated = round(uncertainty, -int(np.floor(np.log10(abs(uncertainty)))))
        # Truncar el coeficiente al mismo número de cifras significativas que la incerteza
        coefficient_truncated = round(coeff, -int(np.floor(np.log10(abs(uncertainty)))))
        print(f"Coeficiente {degree - i}: {coefficient_truncated:.2e} +/- {uncertainty_truncated:.2e}")
    print()



import sympy as sp

def calcular_incertidumbre(funcion, variables, incertidumbres):

    # Calcula las derivadas parciales de la función con respecto a cada variable
    derivadas = [sp.diff(funcion, variable) for variable in variables]

    # Calcula la incertidumbre utilizando el método de propagación de incertidumbres
    delta_f = sp.sqrt(sum([(derivada * incertidumbre)**2 for derivada, incertidumbre in zip(derivadas, incertidumbres)]))

    return delta_f

def calcular_gradiente(funcion, variables):

    # Calcula las derivadas parciales de la función con respecto a cada variable
    gradientes = [sp.diff(funcion, variable) for variable in variables]

    return gradientes


def my_equation_sum(x, y, a, b, num_terms, alpha1, alpha2, alpha5, beta1, beta2, beta5, V4):
    result = 0
    pi = sp.pi
    for n in range(1, num_terms + 1):
        lambda_1 = n * pi / b
        lambda_2 = n * pi / a

        parte_1 = (sp.sin(lambda_1 * y) * sp.sinh(lambda_1 * (x - a))) / sp.sinh(lambda_1 * a)
        parte_2 = (alpha5 * y + beta5) * (sp.sin(lambda_1 * y) * sp.sinh(lambda_1 * x)) / sp.sinh(lambda_1 * a)
        parte_3 = (alpha2 * x + beta2) * (sp.sin(lambda_2 * x) * sp.sinh(lambda_2 * y)) / sp.sinh(lambda_2 * b)
        parte_4 = -(alpha1 * x + beta1) * (sp.sin(lambda_2 * x) * sp.sinh(lambda_2 * (y - b))) / sp.sinh(lambda_2 * b)

        termino_1 = ((-2 * V4 / pi) * ((-1) ** n - 1) / n) * parte_1
        termino_2 = ((2/ pi) * (1 - (-1) ** n) / n) * parte_2
        termino_3 = ((2 / pi) * (1 - (-1) ** n) / n) * parte_3
        termino_4 = ((2 / pi) * (1 - (-1) ** n) / n) * parte_4

        result += - termino_1 + termino_2 + termino_3 + termino_4

    return result


def calculate_sum_until_error(x_val, y_val, a_val, b_val, alpha1_val, alpha2_val, alpha5_val, beta1_val, beta2_val, beta5_val, V4_val, target_error):
    num_terms = 1
    sum_n_minus_1 = my_equation_sum(x_val, y_val, a_val, b_val, num_terms, alpha1_val, alpha2_val, alpha5_val, beta1_val, beta2_val, beta5_val, V4_val)
    sum_n = my_equation_sum(x_val, y_val, a_val, b_val, num_terms + 1, alpha1_val, alpha2_val, alpha5_val, beta1_val, beta2_val, beta5_val, V4_val)
    error = abs(sum_n - sum_n_minus_1)

    while error > target_error:
        num_terms += 1
        sum_n_minus_1 = sum_n
        sum_n = my_equation_sum(x_val, y_val, a_val, b_val, num_terms + 1, alpha1_val, alpha2_val, alpha5_val, beta1_val, beta2_val, beta5_val, V4_val)
        error = abs(sum_n - sum_n_minus_1)

    return sum_n, num_terms


from sklearn.linear_model import LinearRegression

# Generar valores de x
x = np.linspace(0.1, 1.5, 15)[:, np.newaxis]

# Datos de las variables dependientes
V1 = [4.90, 4.74, 4.55, 4.42, 4.31, 4.15, 4.05, 3.87, 3.76, 3.62, 3.52, 3.45, 3.23, 3.12, 2.97]
V2 = [5.03, 4.84, 4.68, 4.49, 4.40, 4.23, 4.10, 4.00, 3.87, 3.64, 3.50, 3.43, 3.27, 3.11, 2.92]
V5 = [2.97, 3.00, 2.96, 2.95, 2.94, 3.00, 2.92]

# Ajustar modelos de regresión lineal para V1 y V2
model1 = LinearRegression().fit(x, V1)
model2 = LinearRegression().fit(x, V2)

# Seleccionar solo los primeros 7 puntos de V5
V5 = V5[:7]
x_V5 = np.linspace(0.1, 1.5, 7)[:, np.newaxis]

# Ajustar modelo de regresión lineal para V5
model3 = LinearRegression().fit(x_V5, V5)

# Generar valores de predicción para V1 y V2
x_pred = np.linspace(0.1, 1.5, 100)[:, np.newaxis]
y_pred1 = model1.predict(x_pred)
y_pred2 = model2.predict(x_pred)

# Generar valores de predicción para V5
x_pred_V5 = np.linspace(0.1, 1.5, 100)[:, np.newaxis]
y_pred3 = model3.predict(x_pred_V5)

# Obtener los coeficientes de las rectas de regresión
alpha1, beta1 = model1.coef_[0], model1.intercept_
alpha2, beta2 = model2.coef_[0], model2.intercept_
alpha5, beta5 = model3.coef_[0], model3.intercept_

# Calcular las incertezas del intercepto utilizando las desviaciones estándar de los errores residuales
residuals1 = V1 - model1.predict(x)
residuals2 = V2 - model2.predict(x)
residuals3 = V5 - model3.predict(x_V5)
sigma1 = np.std(residuals1)
sigma2 = np.std(residuals2)
sigma3 = np.std(residuals3)
n1 = len(V1)
n2 = len(V2)
n3 = len(V5)
beta1_std = sigma1 / np.sqrt(np.sum((x - np.mean(x))**2))
beta2_std = sigma2 / np.sqrt(np.sum((x - np.mean(x))**2))
beta5_std = sigma3 / np.sqrt(np.sum((x_V5 - np.mean(x_V5))**2))
alpha1_std = sigma1 * np.sqrt(1 / n1 + np.mean(x)**2 / np.sum((x - np.mean(x))**2))
alpha2_std = sigma2 * np.sqrt(1 / n2 + np.mean(x)**2 / np.sum((x - np.mean(x))**2))
alpha5_std = sigma3 * np.sqrt(1 / n3 + np.mean(x_V5)**2 / np.sum((x_V5 - np.mean(x_V5))**2))

# Imprimir los valores de alfa y beta con sus incertezas
print("Regression 1 (V1):")
print(f"Alpha (pendiente): {alpha1:.2f} +/- {alpha1_std:.2f}")
print(f"Beta (intercepto): {beta1:.2f} +/- {beta1_std:.2f}")
print()
print("Regression 2 (V2):")
print(f"Alpha (pendiente): {alpha2:.2f} +/- {alpha2_std:.2f}")
print(f"Beta (intercepto): {beta2:.2f} +/- {beta2_std:.2f}")
print()
print("Regression 3 (V5):")
print(f"Alpha (pendiente): {alpha5:.2f} +/- {alpha5_std:.2f}")
print(f"Beta (intercepto): {beta5:.2f} +/- {beta5_std:.2f}")
print()

# Calcular el coeficiente de determinación R^2 para cada modelo
r_squared1 = model1.score(x, V1)
r_squared2 = model2.score(x, V2)
r_squared3 = model3.score(x_V5, V5)

from scipy.stats import pearsonr

# Calcular el coeficiente de correlación para cada par de variables
correlation_coefficient1, _ = pearsonr(x.squeeze(), V1)
correlation_coefficient2, _ = pearsonr(x.squeeze(), V2)
correlation_coefficient3, _ = pearsonr(x_V5.squeeze(), V5)

# Imprimir los coeficientes de correlación
print("Correlation Coefficients (r):")
print(f"V1: {correlation_coefficient1:.2f}")
print(f"V2: {correlation_coefficient2:.2f}")
print(f"V5: {correlation_coefficient3:.2f}")

# Imprimir los coeficientes de determinación R^2
print("Coefficient of Determination (R^2):")
print(f"Regression 1 (V1): {r_squared1:.2f}")
print(f"Regression 2 (V2): {r_squared2:.2f}")
print(f"Regression 3 (V5): {r_squared3:.2f}")

# Graficar los datos y las rectas de regresión
plt.figure(figsize=(10, 6))
plt.scatter(x, V1, color='blue', label='Data 1')
plt.scatter(x, V2, color='red', label='Data 2')
plt.scatter(x_V5, V5, color='green', label='Data 3')
plt.plot(x_pred, y_pred1, color='blue', linestyle='--', label=f'Regression 1: $V1 = {alpha1:.2f}x + {beta1:.2f}$')
plt.plot(x_pred, y_pred2, color='red', linestyle='--', label=f'Regression 2: $V2 = {alpha2:.2f}x + {beta2:.2f}$')
plt.plot(x_pred_V5, y_pred3, color='green', linestyle='--', label=f'Regression 3: $V5 = {alpha5:.2f}x + {beta5:.2f}$')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()
plt.grid(True)
plt.show()

# Ejemplo de uso
x, y = sp.symbols('x y')
variables = [x, y]
a_val = 1.5
b_val = 0.7
num_terms_val = 100
alpha1_val = alpha1
alpha2_val = alpha2
alpha5_val = alpha5
beta1_val = beta1
beta2_val = beta2
beta5_val = beta5
V4_val = 5.03
target_error = 0.00001

print(alpha1, beta1)
print(alpha2, beta2)
print(alpha5, beta5)

# Calcular la función
result = my_equation_sum(x, y, a_val, b_val, num_terms_val, alpha1_val, alpha2_val, alpha5_val, beta1_val, beta2_val, beta5_val, V4_val)

# Calcular el gradiente de la función con respecto a x e y
gradiente_result = calcular_gradiente(result, variables)

print(gradiente_result)
print(f"Número de términos utilizados: {num_terms_val}")


import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define las variables simbólicas
x, y, a, b, alpha1, alpha2, alpha5, beta1, beta2, beta5, V4 = sp.symbols('x y a b alpha1 alpha2 alpha5 beta1 beta2 beta5 V4')

# Define la función simbólica Z
Z_sym = my_equation_sum(x, y, a, b, num_terms_val, alpha1, alpha2, alpha5, beta1, beta2, beta5, V4)

# Convertir la función simbólica en una función que puede evaluar matrices NumPy
Z_func = sp.lambdify((x, y, a, b, alpha1, alpha2, alpha5, beta1, beta2, beta5, V4), Z_sym, 'numpy')

# Definir los rangos para x e y
x_range = np.linspace(0.1-0.05, a_val-0.05, 150)
y_range = np.linspace(0.1-0.05, b_val-0.05, 70)

# Crear una malla para los valores de x e y
X, Y = np.meshgrid(x_range, y_range)

# Calcular los valores de Z utilizando la función numérica
Z = Z_func(X-0.05, Y-0.05, a_val, b_val, alpha1_val, alpha2_val, alpha5_val, beta1_val, beta2_val, beta5_val, V4_val)

# Graficar en 3D
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Graficar la superficie
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_zlim([3, 5])
ax.set_zticks([5, 4.5, 4, 3.5, 3])
ax.set_xlabel('X(cm)', fontsize=14)
ax.set_ylabel('Y(cm)' , fontsize=14)
ax.set_zlabel('', fontsize=14)
ax.set_title('Superficie 3D de la función', fontsize=15)
plt.xticks(size = 11)
plt.yticks(size = 11)


# Agregar barra de color
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()

# Datos de la tabla
data_table = [
    [4.90, 4.74, 4.55, 4.42, 4.31, 4.15, 4.05, 3.87, 3.76, 3.62, 3.52, 3.45, 3.23, 3.12, 2.97],
    [4.91, 4.75, 4.56, 4.43, 4.34, 4.15, 4.04, 3.92, 3.76, 3.61, 3.52, 3.41, 3.22, 3.15, 3.00],
    [4.85, 4.72, 4.57, 4.46, 4.36, 4.22, 4.06, 3.94, 3.84, 3.71, 3.56, 3.41, 3.25, 3.12, 2.96],
    [4.90, 4.74, 4.60, 4.44, 4.32, 4.16, 4.03, 3.89, 3.74, 3.61, 3.50, 3.37, 3.24, 3.10, 2.95],
    [4.92, 4.80, 4.62, 4.50, 4.32, 4.20, 4.03, 3.91, 3.80, 3.63, 3.48, 3.38, 3.26, 3.09, 2.94],
    [4.98, 4.79, 4.64, 4.50, 4.41, 4.24, 4.11, 3.97, 3.86, 3.69, 3.51, 3.38, 3.25, 3.15, 3.00],
    [5.03, 4.84, 4.68, 4.49, 4.40, 4.23, 4.10, 4.00, 3.87, 3.64, 3.50, 3.43, 3.27, 3.11, 2.92]
]

# Definir los rangos para x e y
x_range = np.linspace(0.1-0.05, 1.5-0.05, 15)
y_range = np.linspace(0.1-0.05, 0.7-0.05, 7)

# Crear la malla
X, Y = np.meshgrid(x_range, y_range)
Z = np.array(data_table)

# Convertir la malla en puntos
x_points = X.flatten()
y_points = Y.flatten()
z_points = Z.flatten()

# Graficar los puntos en 3D
ax.set_zlim([3, 5])
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x_points, y_points, z_points, c=z_points, cmap='viridis', marker='o')

ax.set_xlabel('X (cm)',fontsize=14)
ax.set_ylabel('Y (cm)',fontsize=14)
ax.set_zlabel('Z (V)' ,fontsize=14)
ax.set_title('Puntos generados a partir de los datos tabulados', fontsize=15)
plt.xticks(size = 11)
plt.yticks(size = 11)

plt.show()

# Calcular los resultados para cada punto de la malla
calculated_results = []
num_terms_used = []
for i in range(len(x_range)):
    for j in range(len(y_range)):
        x_val = x_range[i]
        y_val = y_range[j]
        result = my_equation_sum(x_val, y_val, a_val, b_val, num_terms_val, alpha1_val, alpha2_val, alpha5_val, beta1_val, beta2_val, beta5_val, V4_val)
        calculated_results.append(result)
        num_terms_used.append(num_terms_val)

# Convertir los resultados calculados en una matriz del mismo tamaño que la tabla de datos
calculated_results = np.array(calculated_results).reshape((len(x_range), len(y_range)))
num_terms_used = np.array(num_terms_used).reshape((len(x_range), len(y_range)))

# Asegurar de que todos los valores en calculated_results sean numéricos
calculated_results = calculated_results.astype(float)



# Calcular la diferencia entre los datos originales y los datos calculados
difference = np.absolute(np.array(data_table) - calculated_results.T)



# Graficar la diferencia entre los datos originales y los datos calculados
print("Diferencia entre los datos originales y los datos calculados:")
print(difference)
plt.figure(figsize=(10, 6))
plt.imshow(difference, cmap='Greys', origin='lower', extent=[0.1, 1.5, 0, 0.7])
#plt.colorbar(label='Diferencia (Original - Calculado)')
#plt.colorbar().set_label(label='V', size=13)
plt.colorbar().ax.tick_params(labelsize=13)
plt.title('Diferencia entre los datos originales y los datos calculados (V)',fontsize=15)
plt.xlabel('X(cm)',fontsize=15)
plt.ylabel('Y(cm)',fontsize=15)
plt.xticks(size = 13)
plt.yticks(size = 13)

plt.show()





# Definir los símbolos para las variables
x, y = sp.symbols('x y')

# Calcular las derivadas parciales de la función simbólica Z_sym con respecto a x e y
dV_dx = sp.diff(Z_sym, x)
dV_dy = sp.diff(Z_sym, y)

# Convertir las derivadas parciales en funciones que pueden ser evaluadas numéricamente
dV_dx_func = sp.lambdify((x, y, a, b, alpha1, alpha2, alpha5, beta1, beta2, beta5, V4), dV_dx, 'numpy')
dV_dy_func = sp.lambdify((x, y, a, b, alpha1, alpha2, alpha5, beta1, beta2, beta5, V4), dV_dy, 'numpy')

# Calcular las componentes del campo eléctrico en cada punto de la malla
Ex = -dV_dx_func(X, Y, a_val, b_val, alpha1_val, alpha2_val, alpha5_val, beta1_val, beta2_val, beta5_val, V4_val)
Ey = -dV_dy_func(X, Y, a_val, b_val, alpha1_val, alpha2_val, alpha5_val, beta1_val, beta2_val, beta5_val, V4_val)

# Ahora Ex y Ey contienen las componentes del campo eléctrico en cada punto de la malla
# Normalizar los vectores
magnitude = np.sqrt(Ex**2 + Ey**2)
Ex_normalized = Ex / magnitude
Ey_normalized = Ey / magnitude

# Ajustar la escala de los vectores
scale = 20  # Ajusta el valor según sea necesario

# Graficar el campo eléctrico usando quiver con vectores normalizados y escala ajustada
plt.figure(figsize=(10, 6))
plt.quiver(X, Y, Ex, Ey, magnitude, cmap='viridis', scale=scale)
plt.xlabel('X (cm)',fontsize=15)
plt.ylabel('Y (cm)',fontsize=15)
plt.title('Campo Eléctrico',fontsize=15)
plt.colorbar().ax.tick_params(labelsize=15)
plt.clim(0, 2)

plt.xticks(size = 15)
plt.yticks(size = 15)

plt.show()


# Datos de la tabla
data_table = np.array([
    [4.90, 4.74, 4.55, 4.42, 4.31, 4.15, 4.05, 3.87, 3.76, 3.62, 3.52, 3.45, 3.23, 3.12, 2.97],
    [4.91, 4.75, 4.56, 4.43, 4.34, 4.15, 4.04, 3.92, 3.76, 3.61, 3.52, 3.41, 3.22, 3.15, 3.00],
    [4.85, 4.72, 4.57, 4.46, 4.36, 4.22, 4.06, 3.94, 3.84, 3.71, 3.56, 3.41, 3.25, 3.12, 2.96],
    [4.90, 4.74, 4.60, 4.44, 4.32, 4.16, 4.03, 3.89, 3.74, 3.61, 3.50, 3.37, 3.24, 3.10, 2.95],
    [4.92, 4.80, 4.62, 4.50, 4.32, 4.20, 4.03, 3.91, 3.80, 3.63, 3.48, 3.38, 3.26, 3.09, 2.94],
    [4.98, 4.79, 4.64, 4.50, 4.41, 4.24, 4.11, 3.97, 3.86, 3.69, 3.51, 3.38, 3.25, 3.15, 3.00],
    [5.03, 4.84, 4.68, 4.49, 4.40, 4.23, 4.10, 4.00, 3.87, 3.64, 3.50, 3.43, 3.27, 3.11, 2.92]
])

# Definir los rangos de x y y
x = np.linspace(0.05, 1.45, 15)
y = np.linspace(0.05, 0.65, 7)

# Calcular las derivadas parciales usando diferencias finitas
delta_x = 0.1
delta_y = 0.1

# Inicializar las matrices para las derivadas
dV_dx = np.zeros_like(data_table)
dV_dy = np.zeros_like(data_table)

# Calcular dV/dx
for i in range(data_table.shape[0]):
    for j in range(data_table.shape[1]):
        if j == 0:  # Diferencia hacia adelante
            dV_dx[i, j] = (data_table[i, j + 1] - data_table[i, j]) / delta_x
        elif j == data_table.shape[1] - 1:  # Diferencia hacia atrás
            dV_dx[i, j] = (data_table[i, j] - data_table[i, j - 1]) / delta_x
        else:  # Diferencia centrada
            dV_dx[i, j] = (data_table[i, j + 1] - data_table[i, j - 1]) / (2 * delta_x)

# Calcular dV/dy
for i in range(data_table.shape[0]):
    for j in range(data_table.shape[1]):
        if i == 0:  # Diferencia hacia adelante
            dV_dy[i, j] = (data_table[i + 1, j] - data_table[i, j]) / delta_y
        elif i == data_table.shape[0] - 1:  # Diferencia hacia atrás
            dV_dy[i, j] = (data_table[i, j] - data_table[i - 1, j]) / delta_y
        else:  # Diferencia centrada
            dV_dy[i, j] = (data_table[i + 1, j] - data_table[i - 1, j]) / (2 * delta_y)

# Calcular la magnitud del campo eléctrico en cada punto
magnitud = np.sqrt(dV_dx**2 + dV_dy**2)

# Graficar los vectores del campo eléctrico con color
plt.figure(figsize=(10, 6))
plt.quiver(x, y, -dV_dx, -dV_dy, magnitud, cmap='viridis', scale=20)
plt.title('Campo eléctrico', fontsize=15)
plt.xlabel('X (cm)', fontsize=15)
plt.ylabel('Y (cm)', fontsize=15)
plt.colorbar().ax.tick_params(labelsize=15)
plt.clim(0, 2)
plt.xticks(size = 15)
plt.yticks(size = 15)

plt.show()

# Datos de incertidumbre
incerteza_x = 0.1
incerteza_y = 0.1
incerteza_voltaje = 0.01


# Calcular la incerteza en las componentes del campo eléctrico
incerteza_ex = round(incerteza_voltaje / delta_x, 1)
incerteza_ey = round(incerteza_voltaje / delta_y, 1)

# Función para truncar las coordenadas
def truncar_coordenada(coordenada):
    return round(coordenada, 1)

# Función para formatear la incerteza
def formato_incerteza(valor):
    if valor == int(valor):
        return int(valor)
    return round(valor, 1)

# Imprimir la incerteza en la tabla
print("Incerteza en las componentes del campo eléctrico (±):")
for i in range(data_table.shape[0]):
    for j in range(data_table.shape[1]):
        print(f"Punto ({truncar_coordenada(x[j])}, {truncar_coordenada(y[i])}): Ex = {dV_dx[i, j]:.1f} ± {formato_incerteza(incerteza_ex)}, Ey = {dV_dy[i, j]:.1f} ± {formato_incerteza(incerteza_ey)}")



# Crear un DataFrame de pandas con los datos y las incertezas
df_list = []

for i in range(data_table.shape[0]):
    for j in range(data_table.shape[1]):
        coordenada_x = truncar_coordenada(x[j])
        coordenada_y = truncar_coordenada(y[i])
        ex = dV_dx[i, j]
        ey = dV_dy[i, j]
        df_list.append(pd.DataFrame({'X': [coordenada_x],
                                     'Y': [coordenada_y],
                                     'Ex': [ex],
                                     'Ey': [ey],
                                     'Incerteza Ex': [incerteza_ex],
                                     'Incerteza Ey': [incerteza_ey]}))

df = pd.concat(df_list, ignore_index=True)


# Guardar el DataFrame en un archivo Excel
df.to_excel('datos_incerteza.xlsx', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Función para calcular el área de la sección transversal
def calcular_area(diametro):
    radio = diametro / 2
    return np.pi * (radio ** 2)

# Función para calcular la resistividad
def calcular_resistividad(longitud, resistencia, area):
    return resistencia * area / longitud

# Cargar datos desde Excel
excel_file = 'libroresistencia.xlsx'  # Reemplaza con el nombre de tu archivo Excel

# Leer todas las hojas del archivo Excel en un diccionario de DataFrames
xls = pd.ExcelFile(excel_file)
sheet_names = xls.sheet_names

# Iterar sobre cada hoja en el archivo Excel
for sheet_name in sheet_names:
    # Leer la corriente desde la celda B2 de la hoja actual
    corriente_mA = pd.read_excel(excel_file, sheet_name=sheet_name, header=None, usecols="B", nrows=2).iloc[1, 0]
    corriente = corriente_mA / 1000  # Convertir corriente de mA a A

    # Leer la hoja actual en un DataFrame, saltando la primera fila que contiene nombres de columnas
    df = pd.read_excel(excel_file, sheet_name=sheet_name, header=1, usecols="A:E", skiprows=1, nrows=21)

    # Obtener datos de longitud, resistencia y diámetro
    longitud = df.iloc[:, 0]  # Primera columna para longitud
    resistencia = df.iloc[:, 3] * 1000  # Cuarta columna para resistencia en kΩ a Ω
    diametro = df.iloc[:, 4] / 1000  # Quinta columna para diámetro en mm a m

    # Calcular el área de la sección transversal
    area = calcular_area(diametro)

    # Calcular la resistividad
    resistividad = calcular_resistividad(longitud, resistencia, area)

    # Calcular el promedio de resistividad y la incertidumbre
    promedio_resistividad = np.mean(resistividad)
    incertidumbre_resistividad = np.std(resistividad, ddof=1) / np.sqrt(len(resistividad))

    # Realizar regresión lineal
    model = LinearRegression()
    X = longitud.values.reshape(-1, 1)  # Convertir a matriz columna
    y = resistencia.values.reshape(-1, 1)
    model.fit(X, y)
    y_pred = model.predict(X)

    # Obtener coeficientes de la regresión lineal
    coeficiente = model.coef_[0][0]
    intercepto = model.intercept_[0]

    # Graficar datos y regresión lineal
    plt.figure(figsize=(8, 6))
    plt.scatter(longitud, resistencia, label=f'Datos reales - {sheet_name}')
    plt.plot(X, y_pred, color='red', linewidth=2, label=f'R = {coeficiente:.2f} * L + {intercepto:.2f}')
    plt.xlabel('Longitud (m)')
    plt.ylabel('Resistencia (Ω)')
    plt.title(f'Regresión lineal de Resistencia vs Longitud - {sheet_name} (Corriente = {corriente_mA:.2f} mA)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Graficar resistividad vs longitud con incertidumbre
    plt.figure(figsize=(8, 6))
    plt.scatter(longitud, resistividad, label=f'Resistividad - {sheet_name}')
    plt.errorbar(longitud, resistividad, yerr=incertidumbre_resistividad, fmt='o', capsize=5, label='Incertidumbre')
    plt.axhline(promedio_resistividad, color='red', linewidth=2, linestyle='--', label=f'Promedio Resistividad = {promedio_resistividad:.2f} Ω·m')
    plt.xlabel('Longitud (m)')
    plt.ylabel('Resistividad (Ω·m)')
    plt.title(f'Resistividad vs Longitud - {sheet_name} (Corriente = {corriente_mA:.2f} mA)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Imprimir la incertidumbre del promedio de resistividad
    print(f"Incertidumbre del promedio de resistividad en {sheet_name}: {incertidumbre_resistividad:.2f} Ω·m")



```
